```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = yf.download("RELIANCE.NS", start="2022-01-01", end="2024-01-01")
df.head()
```

```python
plt.figure()
plt.plot(df['Close'])
plt.title("Reliance Stock Price")
plt.xlabel("Date")
plt.ylabel("Price")
plt.show()
```

```python
df['MA50'] = df['Close'].rolling(window=50).mean()

plt.figure()
plt.plot(df['Close'], label='Close Price')
plt.plot(df['MA50'], label='50-Day Moving Average')
plt.legend()
plt.title("Stock Price with Moving Average")
plt.show()
```

```python
df['Returns'] = df['Close'].pct_change()

plt.figure()
df['Returns'].hist()
plt.title("Daily Returns Distribution")
plt.show()
```

```python
stock1 = yf.download("RELIANCE.NS", start="2022-01-01", end="2024-01-01")
stock2 = yf.download("TCS.NS", start="2022-01-01", end="2024-01-01")

plt.figure()
plt.plot(stock1['Close'], label='Reliance')
plt.plot(stock2['Close'], label='TCS')
plt.legend()
plt.title("Stock Comparison")
plt.show()
```
